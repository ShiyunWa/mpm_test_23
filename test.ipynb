# Notebook to visualize the normal distribution

```python
import numpy as np
import matplotlib.pyplot as plt

def norm(loc, scale, x):
    return 1 / (np.sqrt(2 * np.pi) * scale) * np.exp(- (x - loc)**2 / 2 * scale**2)

features = np.linspace(-5, 5, 100)
targets_1 = norm(loc=0, scale=0.5, x=features)
targets_2 = norm(loc=0, scale=1, x=features)
targets_3 = norm(loc=0, scale=2, x=features)

plt.figure(figsize=(6, 6), dpi=100)
plt.plot(features, targets_1, label='miu = 0, theta_square = 0.5')
plt.plot(features, targets_2, label='miu = 0, theta_square = 1')
plt.plot(features, targets_3, label='miu = 0, theta_square = 2')
plt.legend()
plt.show()
```
